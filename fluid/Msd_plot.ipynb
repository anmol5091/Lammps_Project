```python
import numpy as np
import scipy as sp
```

```python
with open('msd_data.lammps','r') as f:
    msd_data=f.readlines()
```

```python
def get_msd_value(msd_data,i):
    l=msd_data[i].split()
    return float(l[7])
#msd function
def msd(msd_data):
    msd_set=[]
    for i in range(0,290,10):
        base_value=get_msd_value(msd_data,i)
        msd_value=[]
        for j in range(i,i+10):
            msd_value.append(get_msd_value(msd_data,j)-base_value)
        msd_set.append(msd_value)
    avg_msd=[]
    for i in range(0,10):
        sum=0
        for j in range(len(msd_set)):
            sum+=msd_set[j][i]
        avg_msd.append(sum/len(msd_set))
    #graph plotting
    import matplotlib.pyplot as plt
    plt.plot(avg_msd)
    plt.show()
msd(msd_data)
```
