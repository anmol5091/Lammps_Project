```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
with open('nvtdata.rdf.0.3','r') as f:
    system_data=f.readlines()
```

```python
x=[]
y=[]
def get_system_coordinates(system_data,start,end):
    for i in system_data[start:end]:
        m=i.split()
        x.append(float(m[1]))
        y.append(float(m[2]))
get_system_coordinates(system_data,4,103)            
plt.plot(x,y)
plt.show()       
```
